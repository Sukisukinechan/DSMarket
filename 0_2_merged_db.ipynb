```python
## Load data (3.1):
unit_id = pd.read_csv("E:/Capstone-Retail/data_dsmarket/new_files/unit_id.csv") #Marina Windows
#unit_id = pd.read_csv("/Users/marinaramiropareta/Desktop/Capstone_Retail/data_dsmarket/new_files/unit_id.csv") #Marina Mac

unit_id.head()
```

```python
unit_id.shape
```

```python
## Load data (2):
item_sales = pd.read_csv("E:/Capstone-Retail/data_dsmarket/original/item_sales.csv") #Marina Windows
#item_sales = pd.read_csv("/Users/marinaramiropareta/Desktop/Capstone_Retail/data_dsmarket/originals/item_sales.csv") #Marina Mac
item_sales.head()
```

```python
cols = item_sales.columns.values.tolist()
cols_to_drop = cols[7:]
item_sales.drop(cols_to_drop, axis=1, inplace=True)

item_sales.head()
```

```python
sales_calendar = pd.merge(
    left = unit_id,
    right = item_sales,
    how = "left",
    on = ["id"]
)

sales_calendar.head()
```

```python
sales_calendar.shape
```

```python
## Load data (3.2):
date_id = pd.read_csv("E:/Capstone-Retail/data_dsmarket/new_files/date_id.csv") #Marina Windows
#date_id = pd.read_csv("/Users/marinaramiropareta/Desktop/Capstone_Retail/data_dsmarket/new_files/date_id.csv") #Marina Mac
date_id .head()
```

```python
sales_calendar = pd.merge(
    left = sales_calendar,
    right = date_id,
    how = "left",
    on = ["id", "year_week" ]
)

sales_calendar.head()
```

```python
sales_calendar.shape
```

```python
## Load data (2):
item_prices = pd.read_csv("E:/Capstone-Retail/data_dsmarket/new_files/item_prices_mod.csv") #Marina Windows
#item_prices = pd.read_csv("/Users/marinaramiropareta/Desktop/Capstone_Retail/data_dsmarket/new_files/item_prices_mod.csv") #Marina Mac
item_prices.head()
```

```python
item_prices.drop("Unnamed: 0", axis=1, inplace=True)
item_prices.head()
```

```python
sales_calendar_prices = pd.merge(
    left = sales_calendar,
    right = item_prices,
    how = "left",
    on = ["id", "year_week" ]
)

sales_calendar_prices.head()
```

```python
sales_calendar_prices.shape
```

```python
sales_calendar_prices.to_csv("E:/Capstone-Retail/data_dsmarket/new_files/sales_calendar_prices.csv") #Marina Windows
#sales_calendar_prices.to_csv("/Users/marinaramiropareta/Desktop/Capstone_Retail/data_dsmarket/new_files/sales_calendar_prices.csv") #Marina Mac
```
