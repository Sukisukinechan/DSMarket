### Primer descuento

```python
# Cojer el primer valor de cada item por fecha:
last_week = df_final.groupby(["date", "item"]).first()
last_week.reset_index(inplace = True)
```

```python
# Obtener las fechas de hace una semana:
last_week["date_last_week"] = last_week["date"] - timedelta(days = 7 )
last_week
```

```python
columns_to_keep = ["date", "item", "sell_price", "date_last_week"]
columns_to_drop = last_week.columns.difference(columns_to_keep)
last_week.drop(columns=columns_to_drop, inplace=True)

last_week
```

```python
discount = last_week.copy()
discount.drop(columns="date_last_week", inplace=True)
discount
```

```python
# Cambiamos el nombre de dat y sell_price para poder hacer el merge de los precios de hace una semana
discount.rename(columns = {"date": "date_last_week", "sell_price": "sell_price_last_week"}, inplace=True)
discount.head()
```

```python
first_discount = pd.merge(
    left = last_week,
    right = discount,
    how = 'left',
    on = ["date_last_week", "item"])

first_discount
```

```python
# Esto calcula el ratio del precio de venta de la última semana al precio de venta actual. 
# Si este número es positivo, indica un descuento (es decir, una disminución en el precio). 
# Si este número es negativo, indica un incremento en el precio.
first_discount["discount"] = (first_discount["sell_price_last_week"] - first_discount["sell_price"]) / first_discount["sell_price_last_week"]
first_discount
```

```python
# Seleccionar descuentos significativos (5% de descuento) y descartar los demás:
first_discount["discount"] = np.where((first_discount.discount > 0.05) , first_discount.discount, 0)
first_discount = first_discount[(first_discount["discount"]) != 0]
first_discount
```

```python
# Ordenar por fecha y seleccionar solo el primer elemento de descuento para cada item ("first_discount"):
first_discount = first_discount.sort_values(by="date").groupby("item").first()
first_discount.reset_index (inplace = True)
first_discount = first_discount.rename(columns = {"discount": "first_discount"})
first_discount
```

```python
first_discount.drop(columns=["date_last_week","sell_price_last_week"], inplace=True)
first_discount = first_discount.rename(columns = {"date": "date_first_discount", "sell_price" : "sell_price_first_discount"})
first_discount
```

```python
df_cluster_item = pd.merge(
    left = df_cluster_item,
    right = first_discount,
    how = 'left',
    on = ["item"])

df_cluster_item.head()
```

```python
## Ajustar los nulos de las columnas: aquellos productos que han mantenido el precio a lo largo de la ts
df_cluster_item["sell_price_first_discount"] = df_cluster_item["sell_price_first_discount"].fillna(df_cluster_item["initial_sell_price"])
df_cluster_item["first_discount"].fillna(0, inplace=True)
df_cluster_item['date_first_discount'].fillna(df_final['date'].max(), inplace=True)

df_cluster_item.head()
```

### Descuento "on date" (>5% respecto el precio inicial)

El término **descuento "on date"** se refiere generalmente a una reducción de precio u oferta promocional aplicada a un producto o servicio en una fecha específica. Significa que hay un precio con descuento disponible para esa fecha en particular. 
Por ejemplo, un comerciante puede ofrecer un descuento en un día específico, como un día festivo o un evento de ventas especial. Este descuento permite a los clientes comprar el producto a un precio más bajo que el precio regular.
El concepto de descuento en la fecha se utiliza comúnmente en marketing y ventas minoristas para atraer a los clientes y aumentar las ventas durante períodos o eventos específicos. Puede ser una estrategia efectiva para incentivar a los clientes a realizar una compra y crear un sentido de urgencia en torno a la oferta con descuento.

### Otras variables de interés

```python
def calcula_porcentaje(df, columnas):
    for columna in columnas:
        df[columna + '_%'] = (df[columna] / df['units']) * 100
        df = df.drop(columns=[columna])
    return df
```

```python
sales_to_percentage =  ["week_sales",
                        "weekend_sales",
                        "summer_sales",
                        "christmas_sales",

                        "event_sales",
                        "event_SuperBowl_sales",
                        "event_Thanksgiving_sales",
                        "event_Ramadan_sales",
                        "event_NewYear_sales",
                        "event_Easter_sales",

                        "discount_on_date_sales"
                        ]
```

```python
df_cluster_item_final = calcula_porcentaje(df_cluster_item_final, sales_to_percentage)
df_cluster_item_final.head()
```

```python
df_cluster_item_final["mean_pvp"] = df_cluster_item_final["revenue"] / df_cluster_item_final["units"]
df_cluster_item_final["units_per_purchase"] = df_cluster_item_final["units"] / df_cluster_item_final["days_with_purchases"]
df_cluster_item_final["revenue_per_purchase"] = df_cluster_item_final["revenue"] / df_cluster_item_final["days_with_purchases"]
```

```python
df_cluster_item_final["full_price_prediction_temp"] = df_cluster_item_final["initial_sell_price"] * df_cluster_item_final["units"]
df_cluster_item_final["mean_discount"] = (df_cluster_item_final["full_price_prediction_temp"] - df_cluster_item_final["revenue"]) / df_cluster_item_final["full_price_prediction_temp"] 
#P.S.: Si "mean_discount" és positivo, hay descuento.

del(df_cluster_item_final["full_price_prediction_temp"])

df_cluster_item_final.head()
```
